# Problem 1

##### Analysis - 
Created two vectors - one directly , other with random number poission ditribution function. Par() function sets Graphical Parameters like background color. Plot function plots 1 line with x and y axis, for more lines we use 'lines' function. Grid function adds a grid in background and legend function adds legends on top.

```r
sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
sales2<-rpois(12,34)  # random numbers, Poisson distribution, mean at 34, 12 numbers
par(bg="pink") # Set Graphical Parameters like background color 

plot(sales1, col="blue", type="l", ylim=c(0,100), xlab="Month", ylab="Sales" )
title(main="Sales by Month")

lines(sales2, type="o", pch=22, lty=2, col="red")
grid(nx=NA, ny=NULL)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
```

## Running Again

```r
sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
sales2<-rpois(12,34)  # random numbers, Poisson distribution, mean at 34, 12 numbers
par(bg="cornsilk")

plot(sales1, col="blue", type="p", ylim=c(0,100), xlab="Month", ylab="Sales" )
title(main="Sales by Month")

lines(sales2, type="b", pch=22, lty=2, col="red")
grid(nx=NA, ny=NULL)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
```

##### Analysis - 
Read sales data and plotted a barplot using that data. Histogram show frequency of a particular value.Barplot plots each data one after another.

```r
sales<-read.table("salesdata.txt", 
                 header = TRUE)
sales  # to verify that data has been read
hist(as.matrix(sales1), main="Sales Data", ylab= "Frequency", xlab = "Sales",col="green")
hist(as.matrix(sales2), col="blue", add=T)
legend("topright", inset=.0001, c("Sales1","Sales2"), fill=c("green","blue"), horiz=TRUE)
```

### Problem 3: Use boxplot to compare the two sales data. (Discussion: How will you interpret the graph visualization?)

##### Analysis - 
Created two vectors - one directly , other with random number poission ditribution function. Par() function sets Graphical Parameters like background color. Plot function plots 1 line with x and y axis, for more lines we use 'lines' function. Grid function adds a grid in background and legend function adds legends on top.

```r
fn<-boxplot(sales,col=c("orange","green"))$stats

text(1.45, fn[3,2], paste("Median =", fn[3,2]), adj=0.4, cex=0.7)
text(0.45, fn[3,1],paste("Median =", fn[3,1]), adj=0.1, cex=.5)
grid(nx=NA, ny=NULL)
```
